### VADER Sentiment Analysis

```python
import nltk
from nltk.sentiment import vader
```

```python
sentence = "I normally hate Michael Bay films, but I liked this one."
```

```python
# The SIA is used to determine whether a statement is positive or negative.
sia = vader.SentimentIntensityAnalyzer()
sia.polarity_scores(sentence)
```

```python
# This can be called to see the entire lexicon nltk uses for SIA.
sia.make_lex_dict()
```

```python
# Shortcut method for checking polarity.
def polarity(sentence):
    sia = vader.SentimentIntensityAnalyzer()
    return sia.polarity_scores(sentence)
```

```python
polarity("The book was good.")
```

```python
polarity("The book was GOOD") # Capitalization affects the score.
```

```python
polarity("The book was good") # Lack of punctuation doesn't necessarily have impact...
```

```python
polarity("The book was GOOD!!") # ...although using exclaimations does.
```

```python
polarity("The book was not GOOD!!") # Modifiers flip the score.
```

```python
polarity("The book was good lol") # Colloquialisms tend to increase magnitude as well.
```
